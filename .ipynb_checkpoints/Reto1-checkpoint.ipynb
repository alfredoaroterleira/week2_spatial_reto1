## Paso 2: Unimos los mapas

```python
#Siempre teniendo en cuenta la superposici√≥n
base = countries.plot(facecolor="white",
                      edgecolor='black',
                      linewidth=0.1,
                      figsize=(12,12))

rivers.plot(edgecolor='blue', linewidth=0.4,
            ax=base)
cities.plot(marker='.', color='red', markersize=1,alpha=0.7,
            ax=base)
```

## Paso 3: Definimos el pa√≠s --> Alemania

```python
#Filtramos el pa√≠s
germany=countries[countries.COUNTRY=='Germany']
```

```python
#Podemos crear otro subconjunto mediante recortes, ya que no todas las datas poseen los mismos campos para filtrar
citiesGermany_clipped = gpd.clip(gdf=cities,
                          mask=germany)
riversGermany_clipped = gpd.clip(gdf=rivers,
                               mask=germany)
```

### Ploteamos

```python
base = germany.plot(facecolor="yellow", edgecolor='black', linewidth=0.4,figsize=(5,5))
citiesGermany_clipped.plot(marker='+', color='red', markersize=15,
                    ax=base)
riversGermany_clipped.plot(edgecolor='blue', linewidth=0.5,
                    ax=base)
```

### Comprobamos la geometr√≠a

```python
germany.geom_type
```

```python
citiesGermany_clipped.geom_type
```

```python
riversGermany_clipped.geom_type
```

## Paso 4: Proyectamos

```python
#Debemos tener en cuenta que el CRS afecta a los siguientes aspectos: forma, √°rea, escala y direcci√≥n. Asimismo, la mayor√≠a de los mapas poseen un CRS predeterminado: 4326
```

```python
print(germany.crs)
```

```python
#Alemania no es la excepci√≥n. Buscamos mayor informaci√≥n
germany.crs.axis_info
```

En este caso no tiene sentido buscar el centroide, ya que para eso nuestro mapa debe estar proyectado y al observar "latitud" y "longitud" nos damos cuenta que no lo est√°.

### Reproyecci√≥n

El CRS que se tiene que emplear para Alemania es el siguiente: 25832

```python
germany.to_crs(25832).crs.axis_info
```

```python
#Ahora s√≠ buscamos el centroide
germany.to_crs(25832).centroid
```

### Reploteamos

```python
base25832=germany.to_crs(25832).plot()
germany.to_crs(25832).centroid.plot(color='red',ax=base25832)
```

```python
#Procedemos a guardarlo junto al resto de mapas
germany_25832=germany.to_crs(25832)
cities_germany_25832=citiesGermany_clipped.to_crs(germany_25832.crs)
rivers_germany_25832=riversGermany_clipped.to_crs(germany_25832.crs)
```

```python
germany_25832.to_file(os.path.join("data","germanyMaps_25832.gpkg"), layer='country', driver="GPKG")
cities_germany_25832.to_file(os.path.join("data","germanyMaps_25832.gpkg"), layer='cities', driver="GPKG")
rivers_germany_25832.to_file(os.path.join("data","germanyMaps_25832.gpkg"), layer='rivers', driver="GPKG")
```

```python
germany_25832.centroid
```

```python
print(germany_25832.crs)
```

```python
germany_25832.centroid.to_file(os.path.join("data","germanyMaps_25832.gpkg"), layer='centroid', driver="GPKG")
```

## Paso 5: Agregamos la data espacial (aeropuertos)

```python
import pandas as pd
infoairports=pd.read_csv(os.path.join("data","de-airports.csv"))

infoairports.iloc[[0,1,2,3,-4,-3,-2,-1],:] 
```

```python
#Limpiamos nuestra base de datos
infoairports.drop(index=0,inplace=True)
infoairports.reset_index(drop=True, inplace=True)
infoairports.head()
```

```python
#Revisamos las columnas dentro de nuestra data
infoairports.columns
```

```python
#Nos quedamos con las m√°s importantes
keep=['name','type','latitude_deg', 'longitude_deg','elevation_ft','region_name','municipality']
infoairports=infoairports.loc[:,keep]
```

```python
infoairports.info()
```

```python
#Cambiamos el formato de algunas variables
numericCols=['latitude_deg', 'longitude_deg','elevation_ft']
infoairports[numericCols]=infoairports.loc[:,numericCols].apply(lambda x:pd.to_numeric(x))

infoairports.info()
```

### Vamos a plotear

```python
base = germany_25832.plot(color='white', edgecolor='black')

infoairports.plot.scatter(x = 'longitude_deg', y = 'latitude_deg',ax=base)
```

```python
#Nos sale mal, pero podemos mejorarlo
airports=gpd.GeoDataFrame(data=infoairports.copy(),
                 geometry=gpd.points_from_xy(infoairports.longitude_deg,
                                             infoairports.latitude_deg),
                 crs=germany.crs.to_epsg())
```

```python
#Intentamos plotearlo otra vez
base = germany_25832.plot(color='white', edgecolor='black')
airports.plot(ax=base)
```

```python
#Debemos revisar la data de los aeropuertos
type(airports), type(infoairports)
```

```python
#Ahora, intentamos plotearlo con la versi√≥n proyectada
airports_25832=airports.to_crs(25832)

base = germany_25832.plot(color='white', edgecolor='black')
airports_25832.plot(ax=base)
```

A√∫n tenemos problemas, tenemos que seguir revisando

### Verificamos

```python
#Nos aseguramos que inicialmente ambos datasets poseen el mismo CRS
print(germany_25832.crs)
print(airports_25832.crs)
```

```python
#Ahora buscamos asegurarnos que los "bounds" entre Alemania y los aeropuertos coinciden
print(germany_25832.total_bounds)
print(airports_25832.total_bounds)
```

```python
#AJ√Å! Encontramos un problemas, intentamos solucionarlo
airports_in_germany = gpd.sjoin(airports_25832, germany_25832, how="inner", predicate="intersects")

# Verificamos los bounds de los aeropuertos filtrados
print(airports_in_germany.total_bounds)

# Ploteamos los aeropuertos dentro de Alemania
base = germany_25832.plot(color='white', edgecolor='black')
airports_in_germany.plot(ax=base, color='red')
```

```python
#Pero los puntos est√°n muy grandes, los queremos un poco m√°s peque√±os
base = germany_25832.plot(color='white', edgecolor='black')

airports_in_germany.plot(ax=base, color='red', markersize=5) 
```

```python
#Intentamos unirlo con r√≠os y ciudades
base = germany_25832.plot(color='white', edgecolor='black')

#Aeropuertos con puntos peque√±os
airports_in_germany.plot(ax=base, color='lightgreen', markersize=5)

#Ciudades un poco m√°s grandes
cities_germany_25832.plot(marker='+', color='red', markersize=40, ax=base)

#R√≠os de color azul
rivers_germany_25832.plot(edgecolor='blue', linewidth=0.5, ax=base)
```

```python
#Vemos los tipos de aeropuertos
airports_in_germany['type'].value_counts()
```

```python
#Renombramos la columna
airports_in_germany.rename(columns={'type':'kind'},inplace=True)
airports_in_germany['kind'].value_counts()
```
